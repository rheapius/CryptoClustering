```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

---

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
data_scaled= StandardScaler().fit_transform(df_market_data[['price_change_percentage_24h','price_change_percentage_7d',\
                    'price_change_percentage_14d','price_change_percentage_30d','price_change_percentage_60d','price_change_percentage_200d',\
                    'price_change_percentage_1y']])
data_scaled[0:5]
```

```python
# Create a DataFrame with the scaled data
df_data_scaled= pd.DataFrame(data_scaled,
                    columns=['price_change_percentage_24h','price_change_percentage_7d','price_change_percentage_14d','price_change_percentage_30d',\
                             'price_change_percentage_60d','price_change_percentage_200d','price_change_percentage_1y'])

# Copy the crypto names from the original data
df_data_scaled["Crypto_names"]= df_market_data.index
# Set the coinid column as index
df_data_scaled= df_data_scaled.set_index('Crypto_names')

# Display sample data
df_data_scaled.head()
```

---

### Find the Best Value for k Using the Original Data.

```python
# Create a list with the number of k-values from 1 to 11
k= list(range(1,11))
```

```python
# Create an empty list to store the inertia values
inertia_original=[]

# Create a for loop to compute the inertia with each possible value of k
for i in k:
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
    k_model= KMeans(n_clusters=i, random_state=0)
# 2. Fit the model to the data using `df_data_scaled`
    k_model.fit(df_data_scaled)
# 3. Append the model.inertia_ to the inertia list
    inertia_original.append(k_model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data={'k': k, 'inertia': inertia_original}

# Create a DataFrame with the data to plot the Elbow curve
df_elbow= pd.DataFrame(elbow_data)
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
df_elbow.hvplot.line(
    x="k", 
    y="inertia", 
    title="Elbow Curve", 
    xticks=k
) 
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** Best value of K appears to be 3 or 4, as the graph curves at these points

---

### Cluster Cryptocurrencies with K-means Using the Original Data

```python
# Initialise the K-Means model using the best value for k
model = KMeans(n_clusters=3, random_state=0)
```

```python
# Fit the K-Means model using the scaled data
model.fit(df_data_scaled)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled data
kmeans_predictions = model.predict(df_data_scaled) 

# Print the resulting array of cluster values.
kmeans_predictions
```

```python
# Create a copy of the DataFrame
kmeans_predictions_df= df_data_scaled.copy()
```

```python
# Add a new column to the DataFrame with the predicted clusters
kmeans_predictions_df['kmeans_clusters']=kmeans_predictions

# Display sample data
kmeans_predictions_df.head(5)
```

```python
# Create a scatter plot using hvPlot by setting 
kmeans_predictions_df.hvplot.scatter(
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`. 
    x='price_change_percentage_24h',
    y='price_change_percentage_7d',
# Colour the graph points with the labels found using K-Means and 
    by='kmeans_clusters',
# add the crypto name in the `hover_cols` parameter to identify 
    hover_cols= ['Crypto_names'])
# the cryptocurrency represented by each data point.
```

---

### Optimise Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca=PCA(n_components=3) 
```

```python
# Use the PCA model with `fit_transform` to reduce to 
# three principal components.
crypto_pca = pca.fit_transform(df_data_scaled)

# View the first five rows of the DataFrame. 
crypto_pca[:5]
```

```python
# Retrieve the explained variance to determine how much information 
# can be attributed to each principal component.
pca.explained_variance_ratio_ 
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:** About 89% of the total variance is condensed into the 3 principal components.

```python
# Create a new DataFrame with the PCA data.
crypto_pca_df= pd.DataFrame(
# Creating a DataFrame with the PCA data
                crypto_pca, columns= ["PCA1", "PCA2", "PCA3"])
# Copy the crypto names from the original data
crypto_pca_df['crypto_names']= df_data_scaled.index

# Set the coinid column as index
crypto_pca_df=crypto_pca_df.set_index("crypto_names")

# Display sample data
crypto_pca_df.head()
```

---

### Find the Best Value for k Using the PCA Data

```python
# Create a list with the number of k-values from 1 to 11
k= list(range(1,11))
```

```python
# Create an empty list to store the inertia values
inertia_pca=[]

# Create a for loop to compute the inertia with each possible value of k
for i in k:

# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
        k_model= KMeans(n_clusters=i, random_state=0)
# 2. Fit the model to the data using `crypto_pca_df`
        k_model.fit(crypto_pca_df)
# 3. Append the model.inertia_ to the inertia list
        inertia_pca.append(k_model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
pca_elbow_data={'k': k, 'inertia': inertia_pca}

# Create a DataFrame with the data to plot the Elbow curve
df_pca_elbow_data= pd.DataFrame(pca_elbow_data)
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
df_pca_elbow_data.hvplot.line(
    x="k", 
    y="inertia", 
    title="Elbow Curve", 
    xticks=k
) 
```

#### Answer the following questions: 

* **Question:** What is the best value for `k` when using the PCA data?

  * **Answer:** Best value of 'k' using PCA data is 3 or 4 as displayed on the graph


* **Question:** Does it differ from the best k value found using the original data?

  * **Answer:** The value does not differ from the best k value found using original data

### Cluster Cryptocurrencies with K-means Using the PCA Data

```python
# Initialise the K-Means model using the best value for k
model = KMeans(n_clusters=3)
```

```python
# Fit the K-Means model using the PCA data
model.fit(crypto_pca_df)
```

```python
# Predict the clusters to group the cryptocurrencies using the PCA data
crypto_clusters = model.predict(crypto_pca_df)

# Print the resulting array of cluster values.
print(crypto_clusters)
```

```python
# Create a copy of the DataFrame with the PCA data
df_crypto_clusters= crypto_pca_df.copy()

# Add a new column to the DataFrame with the predicted clusters
df_crypto_clusters['crypto_clusters']= crypto_clusters

# Display sample data
df_crypto_clusters.head()
```

```python
# Create a scatter plot using hvPlot by setting 
df_crypto_clusters.hvplot.scatter(
# `x="PC1"` and `y="PC2"`. 
    x='PCA1', y='PCA2',
# Colour the graph points with the labels found using K-Means and 
    by= 'crypto_clusters',
# add the crypto name in the `hover_cols` parameter to identify 
    hover_cols= ['crypto_names']
# the cryptocurrency represented by each data point.
)
```
